```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df.drop('timestamp' , inplace=True , axis=1)
```

```python
df['ignore'].value_counts()
```

```python
df.drop(['ignore' ,'close_time', 'Unnamed: 0'] , inplace=True , axis=1)
```

```python
x = df.drop(['high'], axis=1)
y = df['high']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
models = [
    LinearRegression(),
    Ridge(),
    Lasso(),
    ElasticNet(),
    SGDRegressor(),
    HuberRegressor()
]
```

```python
def Modeling(x_train , x_test ,y_train , y_test , model_name):
    try:
        model = model_name
        model.fit(x_train , y_train)
        print(f'train = {model.score(x_train , y_train)}')
        print(f'test = {model.score(x_test , y_test)}')
    except:
        print('ERROR')
```

```python
for model in models:
    print(model)
    Modeling(x_train , x_test ,y_train , y_test ,model)
    print('*_*_' * 20)
```

```python
pl = make_pipeline(StandardScaler(),LinearRegression())
```

```python
cv = cross_validate(estimator=pl, X=x, y=y, cv=3, scoring='r2', return_train_score=True)
```

```python
cv['train_score'].mean()
```

```python
cv['test_score'].mean()
```

```python
pl.steps
```

```python
pl_params = {
    'linearregression__fit_intercept': [True, False],
    'linearregression__positive': [True, False]
}
```

```python
pl_srch = GridSearchCV(estimator=pl , param_grid= pl_params , cv = 3 , scoring='r2' , return_train_score=True)
```

```python
pl_srch.fit(x , y)
```

```python
pl_srch.best_estimator_
```

```python
pl_srch.best_score_
```

```python
joblib.dump(pl_srch.best_estimator_ , 'btc.h5')
```

```python
df_2 = pd.read_csv('BTCUSDT-trades-2025-06-25.csv')
df_2.columns = ['trade_id', 'price', 'qty', 'quote_qty', 'timestamp', 'is_buyer_maker', 'is_best_match']
df_2['timestamp'] = pd.to_datetime(df_2['timestamp'], unit='us')
df_2['hour'] = df_2['timestamp'].dt.floor('h')

df_2.drop(['is_best_match', 'timestamp', 'trade_id'], axis=1, inplace=True)

df_2['usd_spent'] = df_2['price'] * df_2['qty']

agg_df = df_2.groupby('hour').agg(
    open=('price', 'first'),
    low=('price', 'min'),
    close=('price', 'last'),
    usd_spent=('usd_spent', 'sum'),
    volume=('qty', 'sum'),
    quote_asset_volume=('quote_qty', 'sum'),
    trades=('price', 'count'),
    taker_buy_base=('qty', lambda x: x[df_2.loc[x.index, 'is_buyer_maker'] == False].sum()),
    taker_buy_quote=('quote_qty', lambda x: x[df_2.loc[x.index, 'is_buyer_maker'] == False].sum()),
).reset_index()

open_price = agg_df['open'].iloc[0]
lowest_price = agg_df['low'].min()
close_price = agg_df['close'].iloc[-1]
total_volume = agg_df['volume'].sum()
total_usd_spent = agg_df['usd_spent'].sum()
total_trades = agg_df['trades'].sum()
total_taker_btc = agg_df['taker_buy_base'].sum()
total_taker_usdt = agg_df['taker_buy_quote'].sum()

titles = [
    "Opening Price",
    "Lowest Price of the Day",
    "Closing Price",
    "Total BTC Volume",
    "Total USD/USDT Spent",
    "Total Trades",
    "BTC Bought via Market Buys (Taker)",
    "USDT Spent via Market Buys (Taker)"
]

values = [
    open_price,
    lowest_price,
    close_price,
    total_volume,
    total_usd_spent,
    total_trades,
    total_taker_btc,
    total_taker_usdt
]

for t, v in zip(titles, values):
    print(f"{t}: {v}")
```

```python
%%writefile BTC.py
import streamlit as st
import pandas as pd 
import numpy as np
import sklearn
import category_encoders
import joblib

st.title("Welcome to My BTC Predection for the nest day")
def get_input():
    open = st.number_input("Enter a number of Start price")

    low = st.number_input("Enter a number of Lowest price" )

    close = st.number_input("Enter a number of End price")

    volume = st.number_input("Enter a Total BTC Traded (Volume)")

    quote_asset_volume = st.number_input("Enter a 	Total USDT Traded")

    trades = st.number_input("Enter a number Number of Trades")

    taker_buy_base = st.number_input("Enter a number of BTC Bought via Market Buys (Taker Buy Volume)")

    taker_buy_quote = st.number_input("Enter a number of USDT Spent via Market Buys (Taker Buy Value)")

    year = st.slider("Enter Year", min_value=2025, max_value=2030)

    month = st.slider("Enter month", min_value=0, max_value=12)

    day = st.slider("Enter day", min_value=0, max_value=31)

    return pd.DataFrame(data=[[open , low , close ,volume , quote_asset_volume , trades ,taker_buy_base,taker_buy_quote, year , month ,day]],
        columns=[ 'open' , 'low' , 'close' ,'volume' , 'quote_asset_volume' , 'trades','taker_buy_base','taker_buy_quote', 'year' , 'month' ,'day'])

predct = get_input()

if st.button("🔮 Predict Next Day's BTC Close Price"):       
    pl = joblib.load('btc.h5')
    st.write(pl.predict(predct))
```
